```python
import guppylang
from guppylang import guppy
from guppylang.std.builtins import owned
from guppylang.std.quantum import qubit, measure, cx

guppylang.enable_experimental_features();
```

### Variable not defined

```python
@guppy
def bad(b: bool) -> int:
    if b:
        x = 4
    return x  # x not defined if b is False


bad.compile()  # Raises an error
```

Similarly, variables must have a unique type when they are used.

### Different types

```python
@guppy
def bad(b: bool) -> int:
    if b:
        x = 4


    else:
        x = True
    return int(x)  # x has different types depending on b


bad.compile()  # Raises an error
```

```python
@guppy
def bad(b: bool) -> int:
    if b:
        x = 4
    else:
        x = True
    return int(x)  # x has different types depending on b


bad.compile()  # Raises an error
```

```python
@guppy
def bad(b: bool) -> int:
    x = 4 if b else True
    return int(x)  # x has different types depending on b


bad.compile()  # Raises an error
```

### Copy violation

```python
@guppy
def bad(q: qubit @ owned) -> qubit:
    cx(q, q)
    return q


bad.compile()  # Raises an error
```

```python
@guppy.declare
def foo(q1: qubit) -> None: ...


@guppy.declare
def use(q: qubit @ owned) -> None: ...


@guppy
def test(q: qubit @ owned) -> None:
    use(q)

    foo(q)


test.compile()
```

### Drop violation

```python
@guppy
def bad(q: qubit @ owned) -> qubit:
    tmp = qubit()
    cx(tmp, q)
    #discard(tmp)  # Compiler complains if tmp is not explicitly discarded
    return q


bad.compile()  # Raises an error
```

```python
@guppy.struct
class MyStruct:
    q: qubit


@guppy
def foo(b: bool) -> bool:
    s = MyStruct(qubit())
    if b:
        return measure(s.q)
    return False


foo.compile()
```

### Illegal assignment

```python
@guppy
def outer(x: int) -> int:
    def nested() -> None:
        x += 1  # Mutation of captured variable x is not allowed

    nested()
    return x


outer.compile_function()  # Raises an error
```

```python
@guppy.struct
class MyStruct1:
    x: "MyStruct2"


@guppy.struct
class MyStruct2:
    q: qubit


@guppy.declare
def use(s: MyStruct2 @ owned) -> None: ...


@guppy
def foo(s: MyStruct1 @ owned) -> qubit:
    use(s.x)
    return s.x.q


foo.compile()
```

```python
@guppy
def foo(qs: list[tuple[qubit, bool]] @ owned) -> list[qubit]:
    rs: list[qubit] = []
    for q, b in qs:
        rs += [q]
        if b:
            break
    return rs


foo.compile()
```

```python
@guppy.struct
class MyStruct:
    q: qubit


@guppy.declare
def use(q: qubit @ owned) -> None: ...


@guppy
def test(s: MyStruct, b: bool) -> None:
    if b:
        use(s.q)


test.compile()
```

```python
@guppy
def foo(x: int) -> int:
    y = x + 1

    def bar() -> None:
        def baz() -> None:
            y += 2

        baz()

    bar()
    return y


foo.compile_function()
```
